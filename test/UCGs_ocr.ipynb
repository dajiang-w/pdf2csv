# 2、识别

```python
import os
from PIL import Image
import matplotlib.pyplot as plt
from paddleocr import PaddleOCR, draw_ocr
import numpy as np
```

```python
ocr= PaddleOCR(use_angle_cls = True, use_gpu = False)
def rectocr(img,rect):
    resizenum = 2
    img = img.resize((img.width * resizenum, img.height * resizenum), Image.ANTIALIAS)
    rectn = (rect[0]  * resizenum, rect[1] * resizenum, rect[2] * resizenum, rect[3] * resizenum)
    icrop = img.crop(rectn)
    imgn = pad_img(icrop,20)
    img_np = np.array(imgn)
    # plt.figure()
    # plt.imshow(imgn)
    # plt.xlabel(rect)
    try:
        
        tmp = ocr.ocr(img_np,cls = True)
        # result = tmp[0][1][0]
        result = ''
        for ii in range(len(tmp)):
            result = result + tmp[ii][1][0] # 多行结果合并
        result = result.replace('\n','|')
        result = result.replace(';','。')
        result = result + ';'
        print(rect,'ocr result:',result)
        return result
    except:
        print('ocrError--------------------',rect)
        # plt.figure()
        # plt.imshow(imgn)
        return 'ocrError;'
    
def pad_img(image, padding_num):    # image is an object of image, padding_num 就是要在图像边缘增加几个像素
    iw, ih = image.size # 原图像尺寸
    w,h = iw + 2 * padding_num, ih + 2 * padding_num
    # print('iw:',iw,'ih:',ih)
    # print('w:',w,'h',h,'padding:',padding_num)
    new_image = Image.new('RGB',(w,h),(256,256,256))
    new_image.paste(image,((w - iw)//2,(h - ih) // 2))
    return new_image
```

```python
def getUCG700data(img):
    pname = rectocr(img,(58,73,97,91))
    pid = rectocr(img,(485,72,551,91))
    reportdate = rectocr(img,(530,943,666,962))
    aao = rectocr(img,(116,138,182,150))
    la = rectocr(img,(116,153,182,166))
    lv = rectocr(img,(116,169,182,183))
    ivs = rectocr(img,(116,186,182,200))
    lvpw = rectocr(img,(116,203,182,214))
    ra = rectocr(img,(116,217,182,228))
    rv = rectocr(img,(116,233,182,246))
    pa = rectocr(img,(116,250,182,262))
    fs = rectocr(img,(116,265,182,275))
    ef = rectocr(img,(116,280,182,293))
    mve = rectocr(img,(368,138,428,149))
    aa = rectocr(img,(368,153,428,166))
    ea = rectocr(img,(368,169,428,183))
    lvot = rectocr(img,(368,217,428,228))
    av = rectocr(img,(368,233,428,246))
    pv = rectocr(img,(368,265,428,277))
    csms = rectocr(img,(39,562,693,793)) # 超声描述
    csts = rectocr(img,(30,813,531,933)) # 超声提示
    
    content = pname + pid + reportdate + aao + la + lv + ivs + lvpw + ra \
    + rv + pa + fs + ef + mve + aa + ea + lvot + av + pv + ' ;' + ' ;' + ' ;' + ' ;' + csms + csts + '\n'
    
    txtfile = 'C:/I/files/UCGs/700.txt'
    ff = open(txtfile,'a+',encoding='utf-8')
    ff.write(content)
    print("file is writen")
    
    
    
    
```

```python
rootpath = r'C:\I\files\UCGs\700'
for root, ds, fs in os.walk(rootpath):
    for f in fs:
        print(os.path.join(root,f))
        img = Image.open(os.path.join(root,f))
        getUCG700data(img)
        print('-'*70)
```
